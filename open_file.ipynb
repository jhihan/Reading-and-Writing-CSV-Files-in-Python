Different way to open csv file and its compressed file (gz)
==

Using pandas library (recommended)
--
pandas is a software library written for the Python programming language for data manipulation and analysis. pandas.read_csv can be used to read a comma-separated values (csv) file or its compression file into DataFrame. 

```python
import pandas as pd
```

```python
data_csv = pd.read_csv( 'Iris.csv' )
```

```python
print(data_csv)
```

```python
data_csv_gz = pd.read_csv( 'Iris.csv.gz', compression='gzip'  )
```

```python
print(data_csv_gz)
```

Using the with and open statement
--
We can use with and open statement to open the normal text file. If the file is compressed, the library gzip is needed.

```python
with open('Iris.csv') as csv_file:
    for line in csv_file:
        print(line)
```

```python
import gzip
```

```python
# Please note that the mode argument here is'rt'. 
# If the default setting 'rb' is used, then the control character b and \n will appeer in the beginning and the end of each line 
with gzip.open('Iris.csv.gz','rt') as gz_file:  
    gz_file.readline()
    for line in gz_file:        
        print(line)
```
